```python
import requests
import shutil
import time
import json
from bs4 import BeautifulSoup as bs

for i in range(1,53):
    res = requests.get('http://www.vincent-van-gogh-gallery.org/the-complete-works-{}-24-3-0.html'.format(i))
    soup = bs(res.text,'lxml')
    homepage = 'http://www.vincent-van-gogh-gallery.org/'
    for j in range(0,24):
        page = homepage+soup.select('.dia')[j]['src']
        name  = soup.select('.dia')[j]['src'].split('/')[-1]
        url = page
        res = requests.get(url, stream=True)
        f = open('E:/Jarvis/ProPic/{}'.format(name),'wb')
        res.raw.decode_content = True
        shutil.copyfileobj(res.raw, f)
        time.sleep(2)
        f.close()
```
